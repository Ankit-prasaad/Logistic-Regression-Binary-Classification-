```python
import pandas as pd
import numpy as np
import streamlit as st
import warnings
warnings.filterwarnings('ignore')
import pickle
```

```python
model=pickle.load(open('log11.pkl','rb'))
```

```python
std_scaler= pickle.load(open('std_sca11.pkl','rb'))
```

```python
#### title
st.title('Model Deployment using Logistic Regressiion')
```

```python
# Index(['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
#        'BMI', 'DiabetesPedigreeFunction', 'Age'],
#       dtype='object')
```

```python
# user input function
def user_input_parameters():

    Pregnancies = st.sidebar.number_input(
        'Number of Pregnancies',
        min_value=0,
        max_value=20,
        value=0
    )

    Glucose = st.sidebar.number_input(
        'Glucose Level',
        min_value=0.0
    )

    BloodPressure = st.sidebar.number_input(
        'Blood Pressure',
        min_value=0.0
    )

    SkinThickness = st.sidebar.number_input(
        'Skin Thickness',
        min_value=0.0
    )

    Insulin = st.sidebar.number_input(
        'Insulin Level',
        min_value=0.0
    )

    BMI = st.sidebar.number_input(
        'BMI',
        min_value=0.0
    )

    DiabetesPedigreeFunction = st.sidebar.number_input(
        'Diabetes Pedigree Function',
        min_value=0.0
    )

    Age = st.sidebar.slider(
        'Age',
        min_value=1,
        max_value=100,
        value=25
    )

    # dataframe
    dict1 = {
        'Pregnancies': Pregnancies,
        'Glucose': Glucose,
        'BloodPressure': BloodPressure,
        'SkinThickness': SkinThickness,
        'Insulin': Insulin,
        'BMI': BMI,
        'DiabetesPedigreeFunction': DiabetesPedigreeFunction,
        'Age': Age
    }

    features = pd.DataFrame(dict1, index=[0])

    return features


# user input
df = user_input_parameters()

st.subheader('User Input')
st.write(df)

# scaling
scaled_data = std_scaler.transform(df)

# prediction
pred = model.predict(scaled_data)

# probability
pred_prob = model.predict_proba(scaled_data)

# button
button = st.button('Predict')

if button:

    st.subheader('Prediction Result')

    if pred[0] == 1:
        st.error('Person is Diabetic')
    else:
        st.success('Person is Non-Diabetic')

    st.subheader('Prediction Probability')

    st.write(f'Non-Diabetic Probability: {pred_prob[0][0]:.2f}')
    st.write(f'Diabetic Probability: {pred_prob[0][1]:.2f}')
```
